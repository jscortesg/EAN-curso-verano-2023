# Remoción de NANs

```python
for col in Regalias_proyectos_red_2.columns:
    print(col, "cantidad de NANs: ", Regalias_proyectos_red_2[col].isna().sum(), 
          "porcentaje de NANs: ", Regalias_proyectos_red_2[col].isna().sum()/Regalias_proyectos_red_2.shape[0])
```

```python
def quitarna(df):
    for col in df.columns:
        porcnan = df[col].isna().sum()/df.shape[0]
        if porcnan < 0.1:
            df[col] = df[col].fillna(0)
        elif porcnan >= 0.1:
            df[col] = df[col].fillna(df[col].mean())
            
            
    return df
        
```

```python
quitarna(Regalias_proyectos_red_2)
```

```python
type(Regalias_proyectos_red_2[col].isna().sum()/Regalias_proyectos_red_2.shape[0])
```

```python
Regalias_proyectos_red_3 = quitarna(Regalias_proyectos_red_2)
```

```python
for col in Regalias_proyectos_red_3.columns:
    print(col, "cantidad de NANs: ", Regalias_proyectos_red_3[col].isna().sum(), 
          "porcentaje de NANs: ", Regalias_proyectos_red_3[col].isna().sum()/Regalias_proyectos_red_3.shape[0])
```
